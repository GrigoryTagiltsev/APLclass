```python
# Homework 0, question 2

iris_data = open("iris/iris.data","r")

# This parses the dataset
data = pd.read_csv('iris/iris.data', header=None)
data.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
print(data)


attributes = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].to_numpy # creates a 150*4 matrix with attributes
species = data['species'] # creates a 150-dimentional vector with species
```

```python
# Homework 0, question 3

f = plt.figure()    
f, axes = plt.subplots(nrows = 4, ncols = 4, sharex=True, sharey = True, figsize=(30,30))
 
for i in range(0, 4):
    for j in range(0, 4):
        for species_name, attribute_value in data.groupby("species"):
            if i==j:
                axes[i][j].text(2, 4, data.columns[i], fontsize=30)
            else:
                axes[i][j].scatter(attribute_value.iloc[:,j], attribute_value.iloc[:,i], label = species_name)
                axes[i][j].legend(fontsize=20)

plt.show()
```
